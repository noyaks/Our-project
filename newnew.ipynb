## הסבר ומקרא לתכונות בדאטה

הציון בתכונות האופי בדאטה, נלקח ממבחן שבוגרי תארים בהודו עושים לאחר שסיימו תואר.<br>
חלק במבחן זה עוסק בתכונות אופי.<br>
תכונות אלו נמדדות כדי להבין את ההתאמה של המועמד לסביבת עבודה מסוימת ולדרישות תפקיד ספציפיות.<br>

Conscientiousness (מצפוניות):<br>
מתארת את רמת הארגון, המשמעת העצמית, והשקדנות של האדם.<br>
אנשים עם ציונים גבוהים בתכונה זו נוטים להיות מאורגנים, אחראיים, ותמיד שואפים לבצע את המשימות שלהם בצורה הטובה ביותר.<br>
אנשים עם ציונים נמוכים עשויים להיות יותר ספונטניים ולעיתים חסרי משמעת עצמית.<br>

Agreeableness (הסכמה<br>
מתארת את רמת האכפתיות, האמפתיה והנדיבות של האדם.<br>
אנשים עם ציונים גבוהים בתכונה זו נוטים להיות ידידותיים, שותפי פעולה, ומסייעים לאחרים.<br>
אנשים עם ציונים נמוכים עשויים להיות יותר תחרותיים ופחות אכפתיים לצרכים של אחרים.<br>

Extraversion (מוחצנות):<br>
מתארת את רמת האנרגיה, החברותיות והנטייה לחיפוש ריגושים של האדם.<br>
אנשים עם ציונים גבוהים בתכונה זו נוטים להיות חברתיים, מדברים הרבה, ומחפשים פעילויות חברתיות.<br>
אנשים עם ציונים נמוכים עשויים להיות יותר שקטים, פרטיים ומעדיפים פעילויות אינדיבידואליות.<br>

Neuroticism (נוירוטיות):<br>
מתארת את רמת הרגישות של האדם למתח, חרדה ושליליות.<br>
אנשים עם ציונים גבוהים בתכונה זו נוטים לחוות רגשות שליליים בתדירות גבוהה יותר, כמו חרדה, דיכאון וכעס.<br>
אנשים עם ציונים נמוכים נוטים להיות רגועים, יציבים רגשית, ופחות מושפעים משליליות.<br>


Openness to Experience (פתיחות לחוויות):<br>
מתארת את רמת הדמיון, הסקרנות והרצון לחקור דברים חדשים של האדם.<br>
אנשים עם ציונים גבוהים בתכונה זו נוטים להיות יצירתיים, בעלי דמיון עשיר ופתוחים לרעיונות וחוויות חדשות.<br>
אנשים עם ציונים נמוכים נוטים להיות יותר שמרנים ומעדיפים שגרה וביטחון.<br>

```python
graduation_year_column = 'GraduationYear'
data_2010 = pj[pj[graduation_year_column] == 2010]
num_graduates_2010 = data_2010.shape[0]
print(f"Number of students who graduated in 2010 is: {num_graduates_2010}")
```

```python
data_2010 = pj[pj['GraduationYear'] == 2010]
num_graduates = data_2010.groupby('Specialization').size().reset_index(name='Number of Graduates')
specializations = num_graduates[num_graduates['Number of Graduates'] > 10]['Specialization']
average_salary = data_2010[data_2010['Specialization'].isin(specializations)]
average_salary = average_salary.groupby('Specialization')['Salary'].mean().astype(int).reset_index()
average_salary.columns = ['Specialization', 'Average Salary']
average_salary = average_salary.sort_values(by='Average Salary', ascending=False)
average_salary
```

```python
years = range(2011, 2015)
percentage_changes = []
for year in years: 
    current_year_data = pj[(pj['GraduationYear'] == year) & (pj['Specialization'].isin(relevant_specializations))]
    previous_year_data = pj[(pj['GraduationYear'] == year - 1) & (pj['Specialization'].isin(relevant_specializations))]
    total_students_current = current_year_data.shape[0]
    total_students_previous = previous_year_data.shape[0]
    percentage_change = ((total_students_current - total_students_previous) / total_students_previous) * 100
    percentage_changes.append(percentage_change)
years = pd.Series(years)
plt.figure(figsize=(8, 6))
sns.barplot(x=years, y=percentage_changes, color='skyblue')
for i, percentage in enumerate(percentage_changes):
    plt.text(i, percentage + 0.5, f'{percentage:.1f}%', ha='center', va='bottom', fontsize=10, color='black')
plt.title('Percentage Change in Number of Graduates from Previous Year (2011-2014)')
plt.ylabel('Percentage Change (%)')
plt.show()
```
